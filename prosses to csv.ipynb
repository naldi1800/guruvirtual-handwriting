```python
import os
import fun
import numpy as np
import pandas as pd
from PIL import Image
```

```python

paths = ["Kelas 1 40", "Kelas 2 8", "Kelas 3 46", "Kelas 2 X"]
```

```python
for path in paths:
    
    output_dir_raw = f"RawData/{path}/"
    output_dir = f"Data/{path}"

    # Gather all file names in the directory
    file_names = [f for f in os.listdir(output_dir) if os.path.isfile(os.path.join(output_dir, f))]

    # List to hold each row: [filename, pixel0, pixel1, ...]
    rows = []

    for fname in file_names:
        # Load image and convert to grayscale or RGB array
        img_path = os.path.join(output_dir, fname)
        flat = fun.load_pixel(img_path)
        # print(flat[1])
        # Prepend filename and append to rows
        rows.append([fname] + flat.tolist())
        # rows.append(flat.tolist())

    # Determine number of pixel columns (assumes all images same size & channels)
    num_pixels = len(rows[0]) - 1
    
    # print((max(rows[0]) + min(rows[0]))/2)
    # print(rows[0])
    # Create column names: first column for filename, then pixel_0, pixel_1, ...
    columns = ["filename"] + [f"pixel_{i}" for i in range(num_pixels)]

    # Create DataFrame and save to CSV
    df = pd.DataFrame(rows, columns=columns)
    csv_output = os.path.join(f"{path}.csv")
    df.to_csv(csv_output, index=False)

    print(f"Saved pixel data for {len(rows)} images to {csv_output}")
```
